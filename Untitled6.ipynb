**Hypothesis testing**

Suppose we say that on a average a match duration is more than 800 minutes.
* So our Bold claim is that H1 : match duration $>$ 800
* so our null hypothesis is H0 : match duration is $ \leq $ 800



```python
from scipy.stats import t
```

```python
def t_score(sample_size,sample_mean,pop_mean,sample_std):
  num= sample_mean-pop_mean
  den = sample_std/sample_size**(1/2)
  return num/den
```

```python
sample = data['matchDuration']
mean = sum(sample)/len(sample)
print(mean)
```

```python
import statistics
std = statistics.pstdev(sample)
print(std)
```

```python
sample_size = len(sample)
sample_mean = mean
pop_mean = 800
sample_std = std
```

```python
t=t_score(sample_size,sample_mean,pop_mean,sample_std)
print(t)
```

As here $H_1$ is having $ > $ sign we should perform here

```python
import scipy
ci = 0.95

alpha = 1-ci

t_critical = scipy.stats.t.ppf(1-alpha/2,df=(sample_size-1))

print(t_critical)
```

```python
if(np.abs(t) > t_critical):
  print('Reject Null Hypothesis ')
else:
  print('Fails to reject Null hypothesis')
```
